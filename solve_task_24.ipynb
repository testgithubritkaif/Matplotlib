### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


````python
# code here
df = pd.read_csv('/content/PowerGeneration - PowerGeneration.csv')
df_top_5_stations = df['Power Station'].value_counts().nlargest(5).index.tolist()
df_filtered = df[df['Power Station'].isin(df_top_5_stations)]
df_station_averages = df_filtered.groupby('Power Station')[['Monitored Cap.(MW)', 'Total Cap. Under Maintenace (MW)']].mean()
plt.scatter(df_filtered['Monitored Cap.(MW)'], df_filtered['Total Cap. Under Maintenace (MW)'], c=LabelEncoder().fit_transform(df_filtered['Power Station']), cmap='viridis', alpha=0.6 )
for station in df_station_averages.index:
    plt.axvline(x=df_station_averages.loc[station, 'Monitored Cap.(MW)'], linestyle='--', color='red', label=f'{station} Avg Monitored Cap.')
    plt.axhline(y=df_station_averages.loc[station, 'Total Cap. Under Maintenace (MW)'], linestyle='--', color='blue', label=f'{station} Avg Total Cap. Under Maintenance')
plt.xlabel('Monitored Cap. (MW)')
plt.ylabel('Total Cap. Under Maintenance (MW)')
plt.title('Monitored Cap. vs Total Cap. Under Maintenance for Top 5 Power Stations')
plt.colorbar()
plt.show()
````

### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
# code here
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.abs(X) - np.abs(Y)
ax = plt.subplot(projection="3d")
ax.plot_surface(x,y,Z,cmap='viridis')
plt.show()
````

### `Problem-6` Use Pandas plot functions

* Line plot of closing value of top 5 Stocks in Year 2020.
* Take top 5 stocks based on total turnover in Year 2020

````python
# code here
df = pd.read_csv('/content/nifty-50 - nifty-50.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_2020 = df[df['Date'].dt.year == 2020]
df_top_5_stocks = df_2020.groupby('Symbol')['Turnover'].sum().nlargest(5).index.tolist()
df_filtered = df_2020[df_2020['Symbol'].isin(df_top_5_stocks)]

for symbol in df_top_5_stocks:
    stock_data = df_filtered[df_filtered['Symbol'] == symbol]
    plt.plot(stock_data['Date'], stock_data['Close'], label=symbol)

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Price of Top 5 Stocks in 2020')
plt.legend()
plt.show()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
# code here
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x , y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z,cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
plt.show()
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
# code here
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x , y = np.meshgrid(x,y)
Z = np.tan(np.log2(x**2 + y**2))

# Surface plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,Z,cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot of z = tan(log2(x^2 + y^2))')
plt.show()

# Contour plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.contour(x,y,Z,cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('3D Contour Plot of z = tan(log2(x^2 + y^2))')
plt.show()

# Filled contour plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.contourf(x,y,Z,cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('3D Filled Contour Plot of z = tan(log2(x^2 + y^2))')
plt.show()
````
